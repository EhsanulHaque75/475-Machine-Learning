```python
import os
import cv2
import matplotlib.pyplot as plt

dataset_path = "/content/drive/MyDrive/MangoLeafBD Dataset"


selected_folders = ['Powdery Mildew', 'Anthracnose', 'Healthy', 'Cutting Weevil',
                    'Sooty Mould', 'Bacterial Canker']

def plot_rgb_histograms(image_path, subplot_position):
    img = cv2.imread(image_path)
    if img is not None:

        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)


        r, g, b = cv2.split(img_rgb)


        plt.subplot(6, 3, subplot_position * 3 - 2)
        plt.hist(r.ravel(), bins=256, color='red', alpha=0.7)
        plt.title(f'{selected_folders[subplot_position-1]} - Red Channel')
        plt.xlabel('Pixel Intensity')
        plt.ylabel('Frequency')

        plt.subplot(6, 3, subplot_position * 3 - 1)
        plt.hist(g.ravel(), bins=256, color='green', alpha=0.7)
        plt.title(f'{selected_folders[subplot_position-1]} - Green Channel')
        plt.xlabel('Pixel Intensity')
        plt.ylabel('Frequency')

        plt.subplot(6, 3, subplot_position * 3)
        plt.hist(b.ravel(), bins=256, color='blue', alpha=0.7)
        plt.title(f'{selected_folders[subplot_position-1]} - Blue Channel')
        plt.xlabel('Pixel Intensity')
        plt.ylabel('Frequency')
    else:
        print(f"Error loading image from {image_path}")


plt.figure(figsize=(15, 18))


for i, subfolder in enumerate(selected_folders, start=1):
    subfolder_path = os.path.join(dataset_path, subfolder)
    if os.path.isdir(subfolder_path):

        for file in os.listdir(subfolder_path):
            if file.lower().endswith(('png', 'jpg', 'jpeg')):
                img_path = os.path.join(subfolder_path, file)
                plot_rgb_histograms(img_path, i)
                break

# Display the plot
plt.tight_layout()
plt.show()
```

**image sharpness check**

```python
def calculate_sharpness(image):

    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    laplacian = cv2.Laplacian(gray, cv2.CV_64F)
    variance = laplacian.var()
    return variance

sharpness_values = []

for subfolder in selected_folders:
    subfolder_path = os.path.join(dataset_path, subfolder)
    for file in os.listdir(subfolder_path):
        if file.lower().endswith(('png', 'jpg', 'jpeg')):
            img_path = os.path.join(subfolder_path, file)
            img = cv2.imread(img_path)
            if img is not None:
                sharpness_values.append(calculate_sharpness(img))

# Plot sharpness histogram
plt.figure(figsize=(10, 6))
plt.hist(sharpness_values, bins=50, color='gray', alpha=0.7)
plt.title("Image Sharpness Distribution")
plt.xlabel("Sharpness (Variance)")
plt.ylabel("Frequency")
plt.show()
```

**preprocess**

```python
import cv2
import numpy as np
import os

dataset_path = "/content/drive/MyDrive/MangoLeafBD Dataset"
selected_folders = ['Powdery Mildew', 'Anthracnose', 'Healthy', 'Cutting Weevil',
                    'Sooty Mould', 'Bacterial Canker']

def preprocess_image(image_path):
    img = cv2.imread(image_path)
    if img is not None:
        img_resized = cv2.resize(img, (224, 224))
        img_normalized = img_resized / 255.0
        img_rgb = cv2.cvtColor((img_normalized * 255).astype(np.uint8), cv2.COLOR_BGR2RGB)
        return img_rgb
    return None

for subfolder in selected_folders:
    subfolder_path = os.path.join(dataset_path, subfolder)
    output_folder = os.path.join(dataset_path, f"preprocessed_{subfolder}")
    os.makedirs(output_folder, exist_ok=True)

    for file in os.listdir(subfolder_path):
        if file.lower().endswith(('png', 'jpg', 'jpeg')):
            img_path = os.path.join(subfolder_path, file)
            img_processed = preprocess_image(img_path)

            if img_processed is not None:
                output_path = os.path.join(output_folder, file)
                img_processed_bgr = cv2.cvtColor(img_processed, cv2.COLOR_RGB2BGR)
                cv2.imwrite(output_path, img_processed_bgr)

print("Preprocessing complete.")
```

** split images into train, val, test sets**

```python
import os
import shutil
from sklearn.model_selection import train_test_split

dataset_path = "/content/drive/MyDrive/MangoLeafBD Dataset"
selected_folders = ['Powdery Mildew', 'Anthracnose', 'Healthy', 'Cutting Weevil',
                    'Sooty Mould', 'Bacterial Canker']

output_train_path = "/content/drive/MyDrive/MangoLeafBD Dataset/train"
output_val_path = "/content/drive/MyDrive/MangoLeafBD Dataset/val"
output_test_path = "/content/drive/MyDrive/MangoLeafBD Dataset/test"


os.makedirs(output_train_path, exist_ok=True)
os.makedirs(output_val_path, exist_ok=True)
os.makedirs(output_test_path, exist_ok=True)


def split_images(folder_name, image_list):
    train_images, test_val_images = train_test_split(image_list, test_size=0.4, random_state=42)
    val_images, test_images = train_test_split(test_val_images, test_size=0.5, random_state=42)

    train_folder = os.path.join(output_train_path, folder_name)
    val_folder = os.path.join(output_val_path, folder_name)
    test_folder = os.path.join(output_test_path, folder_name)

    os.makedirs(train_folder, exist_ok=True)
    os.makedirs(val_folder, exist_ok=True)
    os.makedirs(test_folder, exist_ok=True)


    for img in train_images:
        shutil.copy(img, os.path.join(train_folder, os.path.basename(img)))
    for img in val_images:
        shutil.copy(img, os.path.join(val_folder, os.path.basename(img)))
    for img in test_images:
        shutil.copy(img, os.path.join(test_folder, os.path.basename(img)))


for subfolder in selected_folders:
    subfolder_path = os.path.join(dataset_path, subfolder)
    image_list = [os.path.join(subfolder_path, file) for file in os.listdir(subfolder_path) if file.lower().endswith(('png', 'jpg', 'jpeg'))]


    split_images(subfolder, image_list)


def count_images_in_set(path):
    count = 0
    for subfolder in selected_folders:
        subfolder_path = os.path.join(path, subfolder)
        count += len([f for f in os.listdir(subfolder_path) if f.lower().endswith(('png', 'jpg', 'jpeg'))])
    return count


train_count = count_images_in_set(output_train_path)
val_count = count_images_in_set(output_val_path)
test_count = count_images_in_set(output_test_path)

print(f"Training set image count: {train_count}")
print(f"Validation set image count: {val_count}")
print(f"Test set image count: {test_count}")
```

**extract features with VGG16 and train random forest and Decision  Tree**

```python
import os
import numpy as np
import cv2
import time
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix, roc_curve
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
from tensorflow.keras.applications import VGG16
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.models import Model
from tensorflow.keras.layers import GlobalAveragePooling2D
import matplotlib.pyplot as plt


img_size = (224, 224)
dataset_path = "/content/drive/MyDrive/MangoLeafBD Dataset"
output_train_path = "/content/drive/MyDrive/MangoLeafBD Dataset/train"
output_val_path = "/content/drive/MyDrive/MangoLeafBD Dataset/val"
output_test_path = "/content/drive/MyDrive/MangoLeafBD Dataset/test"


base_model = VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

model = Model(inputs=base_model.input, outputs=GlobalAveragePooling2D()(base_model.output))

def preprocess_image(image_path):
    img = cv2.imread(image_path)
    img = cv2.resize(img, img_size)
    img = img_to_array(img)
    img = np.expand_dims(img, axis=0)
    img = img / 255.0  # Normalize
    return img

def extract_features(image_paths):
    features = []
    for img_path in image_paths:
        img = preprocess_image(img_path)
        feature = model.predict(img)
        features.append(feature.flatten())
    return np.array(features)


def prepare_data():
    image_paths_train = []
    image_paths_val = []
    image_paths_test = []
    labels_train = []
    labels_val = []
    labels_test = []

    selected_folders = ['Powdery Mildew', 'Anthracnose', 'Healthy', 'Cutting Weevil',
                        'Sooty Mould', 'Bacterial Canker']
    label_map = {folder: idx for idx, folder in enumerate(selected_folders)}
    le = LabelEncoder()


    for subfolder in selected_folders:
        subfolder_path = os.path.join(output_train_path, subfolder)
        for img_file in os.listdir(subfolder_path):
            if img_file.lower().endswith(('png', 'jpg', 'jpeg')):
                image_paths_train.append(os.path.join(subfolder_path, img_file))
                labels_train.append(subfolder)


    for subfolder in selected_folders:
        subfolder_path = os.path.join(output_val_path, subfolder)
        for img_file in os.listdir(subfolder_path):
            if img_file.lower().endswith(('png', 'jpg', 'jpeg')):
                image_paths_val.append(os.path.join(subfolder_path, img_file))
                labels_val.append(subfolder)


    for subfolder in selected_folders:
        subfolder_path = os.path.join(output_test_path, subfolder)
        for img_file in os.listdir(subfolder_path):
            if img_file.lower().endswith(('png', 'jpg', 'jpeg')):
                image_paths_test.append(os.path.join(subfolder_path, img_file))
                labels_test.append(subfolder)

    labels_train = le.fit_transform(labels_train)
    labels_val = le.transform(labels_val)
    labels_test = le.transform(labels_test)


    X_train_features = extract_features(image_paths_train)
    X_val_features = extract_features(image_paths_val)
    X_test_features = extract_features(image_paths_test)

    return X_train_features, X_val_features, X_test_features, labels_train, labels_val, labels_test, le


X_train_features, X_val_features, X_test_features, y_train, y_val, y_test, label_encoder = prepare_data()


rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
dt_classifier = DecisionTreeClassifier(random_state=42)


def train_classifier(classifier, X_train_features, y_train, model_name):
    start_time = time.time()
    classifier.fit(X_train_features, y_train)
    training_time = time.time() - start_time
    print(f"{model_name} training time: {training_time:.2f} seconds")
    return classifier, training_time


rf_classifier, rf_training_time = train_classifier(rf_classifier, X_train_features, y_train, "Random Forest")

dt_classifier, dt_training_time = train_classifier(dt_classifier, X_train_features, y_train, "Decision Tree")


def evaluate_model(classifier, X_test_features, y_test, model_name):
    y_pred = classifier.predict(X_test_features)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"\n{model_name} - Accuracy: {accuracy * 100:.2f}%")
    print(f"{model_name} - Classification Report:\n", classification_report(y_test, y_pred))


    y_pred_proba = classifier.predict_proba(X_test_features)
    auc_score = roc_auc_score(y_test, y_pred_proba, multi_class='ovr')
    print(f"{model_name} - AUC Score: {auc_score:.2f}")


    cm = confusion_matrix(y_test, y_pred)
    print(f"{model_name} - Confusion Matrix:\n", cm)


    fpr, tpr, _ = roc_curve(y_test, y_pred_proba[:, 1], pos_label=1)
    plt.plot(fpr, tpr, label=f'{model_name} (AUC = {auc_score:.2f})')


evaluate_model(rf_classifier, X_test_features, y_test, "Random Forest")


evaluate_model(dt_classifier, X_test_features, y_test, "Decision Tree")


plt.title('ROC Curves')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.show()
```
